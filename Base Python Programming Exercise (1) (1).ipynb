```python

data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

int_count = 0
str_count = 0
float_count = 0
bool_count = 0
none_count = 0


for item in data:
    if isinstance(item, int) and not isinstance(item, bool):
        int_count += 1
    elif isinstance(item, str):
        str_count += 1
    elif isinstance(item, float):
        float_count += 1
    elif isinstance(item, bool):
        bool_count += 1
    elif item is None:
        none_count += 1

print(f"Integers: {int_count}")
print(f"Strings: {str_count}")
print(f"Floats: {float_count}")
print(f"Booleans: {bool_count}")
print(f"NoneType: {none_count}")
```

```python
data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

total_sum = 0

for item in data:
    if isinstance(item, (int, float)) and not isinstance(item, bool):
        total_sum += item

print(f"The sum of all numeric values in the list is: {total_sum}")
```

```python
data = [2, 3, 'Py', '10', 1, 'SQL', 5.5, True, 3, 'John', None, 7]

str_list = []

for item in data:
    if isinstance(item, str):
        str_list.append(item)

result = '-'.join(str_list)
print(f"Concatenated string: {result}")
```

```python
# i.Use loops to get the answer

list = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

odd_numbers = []

for number in list:
    if number % 2 != 0:
        odd_numbers.append(number)

print(f"Odd numbers using loops: {odd_numbers}")
```

```python
# ii. Use list comprehensions
list = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

odd_numbers = [number for number in list if number % 2 != 0]

print(f"Odd numbers using list comprehensions: {odd_numbers}")
```

```python
# iii.Use lambda function with filter


li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]
odd_numbers = list(filter(lambda x: x % 2 != 0, li))

print(odd_numbers)


```

```python
import math

def calculate_area(shape, **kwargs):
    """
    Calculate the area of different shapes.

    Args:
    shape (str): Name of the shape (square, rectangle, circle).
    **kwargs: Dimensions of the shape.

    Returns:
    float: Area of the shape.

    Raises:
    ValueError: If invalid shape or dimensions are provided.
    """

    # Validate shape
    valid_shapes = ['square', 'rectangle', 'circle']
    if shape.lower() not in valid_shapes:
        raise ValueError("Invalid shape. Supported shapes: square, rectangle, circle")

    # Calculate area based on shape
    if shape.lower() == 'square':
        if 'side' not in kwargs:
            raise ValueError("Please enter 'side' for a square")
        side = kwargs['side']
        if side <= 0:
            raise ValueError("Side length must be positive")
        return side ** 2

    elif shape.lower() == 'rectangle':
        if 'length' not in kwargs or 'width' not in kwargs:
            raise ValueError("Please enter both 'length' and 'width' for a rectangle")
        length = kwargs['length']
        width = kwargs['width']
        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive")
        return length * width

    elif shape.lower() == 'circle':
        if 'radius' not in kwargs:
            raise ValueError("Please enter 'radius' for a circle")
        radius = kwargs['radius']
        if radius <= 0:
            raise ValueError("Radius must be positive")
        return math.pi * (radius ** 2)
```

```python
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Test the function
number = 13
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
def reverse_string(s):
    """Return the reversed string."""
    return s[::-1]

input_string = "abcd123"
reversed_string = reverse_string(input_string)
print(f"Reversed string: {reversed_string}")
```
